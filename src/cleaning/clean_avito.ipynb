```python
df = df.drop_duplicates(subset="id")
print("Apres suppression doublons:", len(df))
```

```python
df["prix"] = pd.to_numeric(df["prix"], errors="coerce")
```

```python
df["surface"] = pd.to_numeric(df["surface"], errors="coerce")
```

```python
df["titre"] = df["titre"].str.lower().str.strip()
df["localisation"] = df["localisation"].str.lower().str.strip()
df["vendeur"] = df["vendeur"].str.lower().str.strip()
```

```python
df["prix"] = df["prix"].fillna(df["prix"].median())
df["surface"] = df["surface"].fillna(df["surface"].median())
df["chambres"] = df["chambres"].fillna(df["chambres"].median())
df["salles_de_bain"] = df["salles_de_bain"].fillna(df["salles_de_bain"].median())
df["etage"] = df["etage"].fillna(df["etage"].median())
```

```python
# créer prix/m²
df["prix_par_m2"] = df["prix"] / df["surface"]
```

```python
print(df.columns)
```

```python
df = df[df["prix"] > 100_000]          # prix minimum realiste
df = df[df["surface"] > 20]            # surface minimum realiste
df = df[df["prix_par_m2"] < 100_000] 
```

```python
df = df[df["surface"] > 0]
df = df[df["surface"] < 1000]
```

```python
df["prix_par_m2"] = df["prix"] / df["surface"]
```

```python
df["prix_par_m2"] = df["prix"] / df["surface"].replace(0, np.nan)
```

```python
df.columns = df.columns.str.lower().str.strip()
```

```python
df = df.dropna(subset=["date"])
```

```python
#Categorie de prix
df["categorie_prix"] = pd.cut(df["prix"], 
    bins=[0, 500000, 1000000, 2000000, 5000000, float("inf")],
    labels=["Tres bas", "Bas", "Moyen", "Eleve", "Tres eleve"])
```

```python
#Categorie de surface
df["categorie_surface"] = pd.cut(df["surface"],
    bins=[0, 50, 80, 120, 200, float("inf")],
    labels=["Studio", "Petit", "Moyen", "Grand", "Tres grand"])
```

```python
#Categorie d'etage
df["type_etage"] = df["etage"].map({
    0: "Rez de chaussee",
    1: "Bas", 2: "Bas", 3: "Bas",
    4: "Moyen", 5: "Moyen", 6: "Moyen",
}).fillna("Haut")
```

```python
#verification des nouvelles colonnes
print(df[["prix", "categorie_prix", "surface", "categorie_surface", "etage", "type_etage"]].head(10))
```
